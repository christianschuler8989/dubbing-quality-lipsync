```python
# vowels is a list of all vowel symbols in either Praat notation (e.g., "\as") or in Unicode. 
# It is used by Interval methods containsvowel() and startswithvowel(), 
# so changing it, for example, adding new symbols to it or removing symbols 
# used for other purposes in a specific case, will change how those methods function.
textgrids.vowels
```

```python
# diacritics is a dict of all diacritics in Praat notation (as keys) 
# and their Unicode counterparts (as values).

# inline_diacritics and index_diacritics are subsets of diacritics. 
# The former are semantically diacritics but appear as inline symbols, 
# the latter are the “true” diacritics (i.e., under- or overstrikes) 
# that need special handling when transcoding.
textgrids.diacritics
```

```python
def char_range(c1, c2):
# """Generates the characters from `c1` to `c2`, inclusive."""
    for c in range(ord(c1), ord(c2)+1):
        yield chr(c)

```
